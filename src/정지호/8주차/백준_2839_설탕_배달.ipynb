```python
N = int(input()) # 배달할 설탕 무게

if N % 5 ==0:    # 1. N이 5의 배수인 경우
  print(N//5)
else:
  answer=0
  while N > 0:
    N -= 3
    answer+=1
    if N % 5 == 0:    # 2. N이 3과 5로 조합할 수 있을 경우 (11, 18..)
      answer += N//5
      print(answer)
      break
    
    elif N==1 or N==2:   # 3. N이 3과 5로도 못 만들어질 경우 (4..)
      print(-1)
      break

    elif N == 0:       # 4. N이 3의 배수일때 (6, 9....)
      print(answer)
      break
```

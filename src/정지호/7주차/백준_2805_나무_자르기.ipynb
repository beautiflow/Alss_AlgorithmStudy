```python
N, M = map(int, input().split())
li_1 = list(map(int, input().split())) 
start, end = 0, max(li_1) #   : 0   /   :   


#  (  ,     )

while start <= end: #      
  mid = (start + end) // 2
  sum = 0 # mid         

  for i in li_1: # for  sum 
    if i > mid:
      sum += (i-mid)

# sum M        , sum M       .
#   !

  if sum>=M:
    start = mid + 1

  else:
    end = mid - 1

print(end) # start end    end  

```

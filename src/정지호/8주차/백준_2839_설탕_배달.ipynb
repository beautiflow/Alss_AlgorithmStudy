```python
N = int(input()) #   

if N % 5 ==0:    # 1. N 5  
  print(N//5)
else:
  answer=0
  while N > 0:
    N -= 3
    answer+=1
    if N % 5 == 0:    # 2. N 3 5     (11, 18..)
      answer += N//5
      print(answer)
      break
    
    elif N==1 or N==2:   # 3. N 3 5    (4..)
      print(-1)
      break

    elif N == 0:       # 4. N 3  (6, 9....)
      print(answer)
      break
```

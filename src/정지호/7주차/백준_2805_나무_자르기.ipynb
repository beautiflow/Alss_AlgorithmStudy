```python
N, M = map(int, input().split())
li_1 = list(map(int, input().split())) 
start, end = 0, max(li_1) # 절단기 높이 최솟값: 0   /   최댓값: 최대 나무 길이


# 이분탐색 (차례대로 찾아나가는게 아니라, 가운데 값을 통해 찾아나가는 것)

while start <= end: # 최솟값이 최댓값보다 커지기 전까지 계속 반복
  mid = (start + end) // 2
  sum = 0 # mid 높이로 절삭기를 설정할 때 얻는 총 나무 길이 

  for i in li_1: # for 문으로 sum 구해주기
    if i > mid:
      sum += (i-mid)

# sum이 M보다 크거나 같으면 절삭기 높이를 더 높게 설정해야 하고, sum이 M보다 작으면 절삭기 높이가 더 낮게 설정되어야 한다.
# 이를 이분 탐색으로!

  if sum>=M:
    start = mid + 1

  else:
    end = mid - 1

print(end) # start가 end 값보다 커졌을 때의 end 값 출력

```
